```python
def doSearch(start, end, terms):
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch(terms)\
                                           .setSince(start)\
                                           .setUntil(end)\
                                           .setMaxTweets(500)
    return got.manager.TweetManager.getTweets(tweetCriteria)
```

```python
def doSave(tweets):
    tfile = open('extracted-tweets.txt','a', encoding="utf-8")

    attrs = ['username', 'to', 'text', 'retweets', 'favorites', 'replies', 'id', 'permalink', 'author_id', 'date', 'formatted_date', 'mentions', 'hashtags', 'geo', 'urls']
    tfile.write('|'.join(attrs)+'\n')

    for tweet in tweets:

        values = []
        for attr in attrs:
            value = tweet.__dict__[attr]
            values.append(str(value).replace('|', ' '))

        tfile.write('|'.join(values)+'\n')

    tfile.close()
```

```python
def generateResults(list, terms):
    ini = list[0]
    for fin in list[1:-2]:
        
        print("Searching from {} to {}".format(ini.strftime("%Y-%m-%d"), fin.strftime("%Y-%m-%d")))
        result = doSearch(ini.strftime("%Y-%m-%d"), fin.strftime("%Y-%m-%d"), terms)
        
        print("Results {}".format(len(result)))
        doSave(result)
        
        ini = fin
```

```python
def doSearchTerms(startTxt, endTxt, terms):
    start = datetime.datetime.strptime(startTxt, "%Y-%m-%d")
    end = datetime.datetime.strptime(endTxt, "%Y-%m-%d")
    date_generated = [start + datetime.timedelta(days=x) for x in range(0, (end-start).days)]
    
    generateResults(date_generated, terms)
```

```python
def createTerms():
    terms = "#Temblor #Sismo #Terremoto"
    terms = ' '.join(list(set(terms.split(' ')))).strip()
    doSearchTerms("2019-05-16","2019-06-01", terms)
```
