```python
testRT = "https://twitter.com/i/activity/retweeted_popup?id=1106175997160894466"
```

```python
response = urllib.request.urlopen(testRT).read()
```

```python
d = json.loads(response)
```

```python
finalHTML = d['htmlUsers'].replace('\n','').replace(' ',' ')
```

```python
parsed_html = BeautifulSoup(finalHTML)
allusers = parsed_html.body.findAll('span', attrs={'class':'username u-dir u-textTruncate'})
```

```python
users = set()
for user in allusers:
    users.add(user.text)
print(len(users), users )
```

```python
rters = open('rters.txt','w', encoding="utf-8")
rters2 = open('rters2.txt','w', encoding="utf-8")

for user in users:
    print("Check ",user) 
    tweetCriteria = got.manager.TweetCriteria().setUsername(user)\
                                               .setSince("2019-02-05")\
                                               .setUntil("2019-03-21")\
                                               .setMaxTweets(100)
    tweets = got.manager.TweetManager.getTweets(tweetCriteria)
    for tweet in tweets:
        
        #to_zone = tz.gettz('America/Guayaquil')
        #to_zone = tz.tzlocal()
        #to_zone = tz.gettz('Africa/Windhoek')
        #.astimezone(to_zone)
       
        date = tweet.date.strftime("%I:%M - %d %b. %Y").lower()
        txt = tweet.id+' ( '+date+' ) '+tweet.username+' '+tweet.text+'\n'
        
        if 'https://twitter.com/CynthiaViteri6' in tweet.text:
            rters.write(txt)
        else:
            rters2.write(txt)

rters.close()
rters2.close()
```
